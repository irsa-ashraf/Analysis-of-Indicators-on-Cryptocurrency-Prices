```python
import pandas as pd
import numpy as np
```

```python
dogecoin = pd.read_csv("../data/doge_daily_modified.csv")
```

```python
dogecoin.tail(20)
```

```python
dogecoin = dogecoin.drop(columns=['Unnamed: 0'])
dogecoin.head()
```

```python
# Create a daily change percent
dogecoin['Daily_Change_Perc'] = dogecoin["Daily_Change"] / dogecoin["Open"]
```

```python
# Weekly change
dogecoin['Weekly_Change'] = dogecoin['Close'] - dogecoin['Open'].shift(7)
dogecoin['Weekly_Change_Perc'] = dogecoin['Weekly_Change'] / dogecoin['Open'].shift(7)
```

```python
delta = dogecoin['Close'].diff()
up = delta.clip(lower=0)
down = -1*delta.clip(upper=0)
ema_up = up.ewm(com=13, adjust=False).mean()
ema_down = down.ewm(com=13, adjust=False).mean()
rs = ema_up/ema_down
dogecoin['RSI'] = 100 - (100/(1 + rs))
```

```python
def get_streak(df):
    streak = 0
    for row in df.itertuples():
        ind = df.loc[row.Index, 'Daily_Change_Ind']
        if ind == 0 and streak > 0:
            streak = 0
        if ind == 1 and streak < 0:
            streak = 0
        streak += 1 if ind == 1 else -1
        df.loc[row.Index, 'streak'] = streak
```

```python
get_streak(dogecoin)
```
