```python
import pandas as pd
import numpy as np
```

```python
bitcoin = pd.read_csv("../data/bitcoin_daily.csv")
```

```python
bitcoin_tweets = pd.read_csv("../data/bitcoin_daily_tweets.csv")
```

```python
bitcoin['Change'] = bitcoin["Close"] - bitcoin["Open"]
```

```python
bitcoin['Increase'] = np.where(bitcoin['Change'] >= 0, 1, 0)
```

```python
bitcoin['Change_Day_Before'] = bitcoin["Change"].shift(1)
bitcoin['Increase_Day_Before'] = bitcoin["Increase"].shift(1)
```

```python
bitcoin_tweets.head()
```

```python
bitcoin_data = bitcoin.merge(bitcoin_tweets, left_on="Date", right_on="date", how="left")
```

```python
bitcoin_data.head()
```

```python
bitcoin_data['tweets_day_before'] = bitcoin_data['cashtags'].shift(1)
bitcoin_data['sentiment_day_before'] = bitcoin_data['compound'].shift(1)
```
