```python
import pandas as pd
import numpy as np
```

```python
# This dataset is not included in github and can be found here: 
# https://www.kaggle.com/datasets/fabioturazzi/cryptocurrency-tweets-with-sentiment-analysis
tweets = pd.read_csv("tweets_sentiment.csv")
```

```python
tweets.dtypes
```

```python
# eliminate unnecessary fields
tweets_filtered = tweets[['date', 'mentions', 'replies_count', 'retweets_count', 'likes_count', \
                          'hashtags', 'cashtags', 'neg', 'neu', 'pos', 'compound']]
```

```python
tweets_filtered.sample(10)
```

```python
len(tweets_filtered)
```

```python
bitcoin_tweets = tweets_filtered[tweets_filtered["hashtags"].str.contains("dogecoin|doge", regex=True)]
```

```python
len(bitcoin_tweets)
```

```python
bitcoin_tweets.head()
```

```python
bitcoin_tweets["compound_times_retweets"] = bitcoin_tweets["compound"] * bitcoin_tweets["retweets_count"]
bitcoin_tweets["pos_times_retweets"] = bitcoin_tweets["pos"] * bitcoin_tweets["retweets_count"]
bitcoin_tweets["neg_times_retweets"] = bitcoin_tweets["neg"] * bitcoin_tweets["retweets_count"]
```

```python
bitcoin_tweets.head()
```

```python
bitcoin_pivot = bitcoin_tweets.pivot_table(index="date", \
                                           values=["cashtags", "replies_count", "retweets_count",\
                                                   "likes_count", "neg", "neu", "pos", "compound", \
                                                  "compound_times_retweets","pos_times_retweets", \
                                                   "neg_times_retweets"],
                                          aggfunc={"cashtags" : "count", "replies_count" : np.sum,\
                                                   "retweets_count": np.sum, "likes_count": np.sum,\
                                                   "neg": np.mean, "neu": np.mean, "pos": np.mean,\
                                                   "compound": np.mean,  "compound_times_retweets": np.sum,\
                                                   "pos_times_retweets": np.sum, "neg_times_retweets": np.sum})
```

```python
bitcoin_pivot.rename(columns={'cashtags': "count"}, inplace=True)
bitcoin_pivot["pos_weighted"] = bitcoin_pivot["pos_times_retweets"] / bitcoin_pivot['retweets_count']
bitcoin_pivot["neg_weighted"] = bitcoin_pivot["neg_times_retweets"] / bitcoin_pivot['retweets_count']
bitcoin_pivot["compound_weighted"] = bitcoin_pivot["compound_times_retweets"] / bitcoin_pivot['retweets_count']
```

```python
bitcoin_pivot.sample(10)
```

```python
rolling_avg_col = ['count', 'replies_count', 'retweets_count', 'likes_count', 'compound_weighted', 'pos_weighted', 'neg_weighted']
```

```python
for col in rolling_avg_col:
    bitcoin_pivot[f"{col}_avg7"] = bitcoin_pivot[col].rolling(7).mean()
    bitcoin_pivot[f'{col}_daily_diff'] = bitcoin_pivot[col] - bitcoin_pivot[col].shift(1)
    bitcoin_pivot[f'{col}_weekly_diff'] = bitcoin_pivot[col] - bitcoin_pivot[f"{col}_avg7"]
```

```python
bitcoin_pivot.head(10)
```

```python
bitcoin_pivot.to_csv("../data/dogecoin_daily_tweets.csv")
```
